```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = sns.load_dataset('tips')

# Get a concise summary of the dataframe
print("--- .info() ---")
df.info()

# Get descriptive statistics for numerical columns
print("\n--- .describe() ---")
print(df.describe())

# Get counts of unique values for a categorical column
print("\n--- .value_counts() for 'day' ---")
print(df['day'].value_counts())
```

```python
sns.histplot(df['total_bill'], kde=True, bins=20)
plt.title('Distribution of Total Bill Amounts')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex')
plt.title('Total Bill vs. Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```python
sns.boxplot(x='day', y='tip', data=df)
plt.title('Distribution of Tips by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Tip ($)')
plt.show()
```

```python
# Select only numeric columns for the correlation matrix
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
sns.pairplot(df, hue='smoker')
plt.suptitle('Pairplot of Numerical Features by Smoker Status', y=1.02)
plt.show()
```

# Based on this quick EDA, we've identified several key trends:

Primary Driver for Tips: The total_bill is the single biggest factor influencing the tip amount.

Business Patterns: The restaurant is busiest on weekends (Saturday and Sunday), which also see the highest variability in tips.

Group Size Matters: The size of the dining party is correlated with both the bill and the tip, which is expected.

No Strong Gender/Smoking Effect: While there are minor differences, neither gender nor smoking status appears to be a dominant factor in tipping behavior when viewed against the bill amount.
